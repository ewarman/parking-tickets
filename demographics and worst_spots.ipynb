```python
import math
def do_round(numbers, n):
    # n=number of sig digits.
    v = 10**n
    return str(math.ceil(int(numbers) / v) * v)
    
def collapse_digits(address, n):
    return re.sub(r'([0-9]+)',
           lambda x: do_round(x.groups()[0], n),
           address)
```

```python
%matplotlib inline

from collections import Counter
import matplotlib.pyplot as plt
import matplotlib.dates as dts
from datetime import datetime

def produce_insights(tickets,address):
    useless_fields = ['location1','location3','location4','infraction_description','province','tag_number_masked']
    for field in useless_fields:
        del tickets[field]
    #print(tickets.head(100))
    print('TICKETS PER DAY',tickets.shape[0]/(365*8))
    
    st_name = address.split(' ',1)[1]
    #counter of addresses
    addresses = Counter(tickets['location2'].tolist())

    street_numbers = sorted([int(x.split(' ')[0]) for x in addresses.keys()])
    street_order=[]
    sn =[]
    for num in street_numbers:
        cur_add = str(num)+' '+st_name
        street_order.append(addresses[cur_add])
        sn.append(num)

    plt.figure(figsize=(15,6))
    plt.plot(sn,street_order)
    plt.title(address+' ADDRESS DISTRIBUTION')
    plt.show()
    
    #counter of infractions
    infractions = Counter(tickets['infraction_code'].tolist())
    print("INFRACTIONS",infractions)

    #cost analysis
    cost = Counter(tickets['set_fine_amount'].tolist())
    print("COST",cost)
    rev = 0
    for key in cost.keys():
        cst = int(key)
        rev += cst*int(cost[key])
    print("DAILY REVENUE",rev/(365*8))
    #plot dates
    def convert_date(date):
        return datetime.strptime(date, '%Y%m%d')

    dates = Counter(tickets['date_of_infraction'].tolist())

    dates_order = sorted(dates.keys())
    count_order=[]
    d_order=[]
    for date in dates_order:
        count_order.append(dates[date])
        d_order.append(convert_date(date))

    plt.figure(figsize=(15,6))
    plt.plot(d_order,count_order)
    plt.title(address+' DATES')
    plt.show()

    def convert_time(time):
        try:
            return datetime.strptime(time, '%H%M')
        except ValueError:
            return None

    #print(convert_time('1100'))

    times = Counter(tickets['time_of_infraction'].tolist())

    times_order = sorted(times.keys())
    co = []
    to = []
    for time in times_order:
        if convert_time(time):
            co.append(times[time])
            to.append(convert_time(time).time())

    plt.figure(figsize=(15,6))
    plt.plot(to,co)
    plt.title(address+' TIMES')
    plt.show()
```
